### 1.3.9.1 Julia のジェネリック関数

※Juliaでは、引数の型や個数の違う同名関数を**多重定義**できる。  
　例えば上記の2関数は個別に定義されている（2つめの関数定義で1つめを上書きしてはいない）。  
※追加例↓

```julia
function hello(str::String)
    println("Hello " * str * "!") # Julia では文字列の連結は `*`
end
```

```julia
hello("Julia")
```

### 1.4.99 Julia のユーザ定義型

※Julia には、Python や Ruby のような*クラス*という概念はない。  
※その代わり、`type`（または `immutable`）キーワードを利用して新しい**型**を定義し、その型に合うように**関数を（多重）定義**する。

```julia
struct Man
    name::String
end

function hello(m::Man)
    println("Hello " * m.name * "!")
end

function goodbye(m::Man)
    println("Good-bye " * m.name * "!")
end
```

```julia
m = Man("David")
```

```julia
hello(m)
```

```julia
goodbye(m)
```
